## 📚 重要说明：Sim2Real Gap

### ❌ 为什么 MuJoCo 训练的模型不能直接用于实体机械臂？

#### 1. **物理差异**
- **摩擦力**: 仿真中的摩擦模型是简化的
- **惯性**: 真实机械臂有质量分布、惯性
- **延迟**: 实体硬件有通信延迟、电机响应延迟
- **柔性**: 真实材料有弹性变形

#### 2. **视觉差异**
- **渲染 vs 真实摄像头**: 仿真渲染完美，真实摄像头有噪声、畸变、曝光问题
- **光照**: 仿真光照可控，真实环境光照复杂多变
- **纹理**: 仿真纹理是近似的，真实物体有细节、反射

#### 3. **环境差异**
- **桌面**: 仿真中是完美平面，真实桌面有不平整、污渍
- **背景**: 仿真背景简单，真实环境复杂
- **物体**: 仿真物体是理想化的，真实物体有重量、材质差异

### ✅ MuJoCo 的正确用途

1. **算法验证**: 快速测试新想法
2. **流程学习**: 熟悉 LeRobot 工作流
3. **预训练**: 作为预训练基础，再用真实数据微调
4. **教学**: 学习机器人学习概念

### 🔄 Sim2Real 迁移策略

如果您想使用仿真数据：

1. **域随机化 (Domain Randomization)**:
   - 随机化光照、颜色、纹理
   - 随机化物理参数（摩擦、质量）
   - 添加视觉噪声

2. **域适应 (Domain Adaptation)**:
   - 使用 GAN 将仿真图像转换为真实风格
   - 特征对齐技术

3. **混合训练**:
   - 仿真数据预训练
   - 真实数据微调

4. **最佳方案**: **直接在真实机械臂上采集数据** ✅

## 📊 MuJoCo vs 实体机械臂对比

| 特性 | MuJoCo 仿真 | 实体 SO101 机械臂 |
|------|-------------|------------------|
| **成本** | 免费 | ~$1000+ |
| **速度** | 可加速/减速 | 实时 |
| **安全** | 完全安全 | 需要小心操作 |
| **物理准确性** | 近似 | 完全真实 |
| **视觉质量** | 渲染（完美） | 真实摄像头（有噪声） |
| **数据采集** | 快速 | 较慢 |
| **模型迁移** | ❌ 不能直接用于实体 | ✅ 可以直接用 |
| **适用场景** | 学习、原型、预训练 | 最终部署、真实应用 |

### 推荐工作流

```
1. MuJoCo 仿真
   ↓ (验证算法)
2. 小规模真实数据测试
   ↓ (确认可行性)
3. 大规模真实数据采集
   ↓ (训练最终模型)
4. 实体机械臂部署
   ↓ (实际应用)
5. 持续改进
```
