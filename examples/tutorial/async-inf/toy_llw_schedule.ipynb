```python
from __future__ import annotations

from dataclasses import dataclass
from typing import Generic, TypeVar

T = TypeVar("T")

@dataclass(frozen=True)
class LWWState(Generic[T]):
    action_step: int
    value: T
    def __or__(self, other: "LWWState[T]") -> "LWWState[T]":
        if other.action_step > self.action_step:
            return other
        return self
```

```python
a = LWWState(action_step=1, value="a")
b = LWWState(action_step=2, value="b")
c = LWWState(action_step=3, value="c")

assert a | a == a
assert ((a | b) | c) == (a | (b | c))
assert a | b == b | a
```
